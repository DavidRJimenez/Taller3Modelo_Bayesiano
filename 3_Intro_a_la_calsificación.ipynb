## 1.) Cree un conjunto de datos

Cree un conjunto de datos con dos características, cuya variable objetivo es binaria y separable.

```python
def create_data(n_samples):
  n_smaples_class = int(n_samples/2)
  X_1 = np.random.randn(n_smaples_class,2)
  X_1 =  2*(X_1 + 2.5)

  X_2 = np.random.randn(n_smaples_class,2)
  X_2 =  2*(X_2 + 0)

  X = np.empty((n_samples,2))

  X[:n_smaples_class,:] = X_1
  X[n_smaples_class:,:] = X_2

  Y = np.zeros(n_samples)
  Y[:n_smaples_class] = 1

  return X,Y
```

```python
X,y = create_data(1000)
plt.plot(X[y==0,0],X[y==0,1],'.')
plt.plot(X[y==1,0],X[y==1,1],'.')
```
